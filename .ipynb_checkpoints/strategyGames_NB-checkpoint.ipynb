To find best possible feature combinations for better user ratings.

# 3. Prepare Data for Consumption

### 3.1 Importing libraries

```python
import pandas as pd
import numpy as np
import datetime

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### 3.2 Meet and Greet data

```python
# Import data from file
df_raw = pd.read_csv('appstore_games.csv')

# Create a copy for exploration purposes
df = df_raw.copy(deep=True)
```

```python
# Preview data
print(df_raw.info())
df_raw.head()
```

### 3.3 The 4 C's of Data Cleaning : Correcting, Completing, Creating, Converting 

```python
# Removing unsused columns : 'URL', 'ID', 'Icon URL'
df.drop(columns = ['URL', 'ID', 'Icon URL'], inplace=True)
```

```python
# Checking Number of null values in each column
print(df.isnull().sum())
```

```python
df.describe(include = 'all')
```

### Clean Data

We can see almost 70% of the 'Subtitle' column is filled with null values. The clear solution here would be to delete this feature but before doing that I would like to check whether their is a Correlation between 'Average User Rating' and 'Subtitle'.
But 'Subtitle' contains textual data in contrast with float data of 'Average User Rating'. So I created a new feature containing no. of words each Subtitle has and then calculated correlation.

```python
df2 = df.copy(deep=True)
```

```python
df2['sub count'] = df2['Subtitle'].str.split().str.len()
```

```python
df2.corr()
```

Here, correlation between 'Average User Rating' and 'sub count' is +0.02 which is very low. So we can safely remove it from the data.

```python
df.drop(columns='Subtitle', inplace=True)
```

We have to convert 'Original Release Date' and 'Current Version Release Date' to correct format(datetime) for future analysis.

```python
df['Original Release Date']= pd.to_datetime(df['Original Release Date'])
df['Current Version Release Date'] = pd.to_datetime(df['Current Version Release Date'])
```

```python
# Convert 'Size' column from Bytes to Megabytes
df['Size'] = round(df['Size']/1000000)
```

```python
# Removing record of games that doesn't have 'Average User Rating'
df3 = df[pd.notnull(df['Average User Rating'])]
```

```python
# Replacing null values in 'In-app Purchases' with 0
df3['In-app Purchases'].fillna(0, inplace=True)
```

```python
df3['Languages'].fillna('EN',inplace=True)
```

```python
# Rechecking missing values
df3.isnull().sum()
```

```python
# Create a column containg time gap b/w latest update and release date
df3['Update Gap'] = df['Current Version Release Date'] - df['Original Release Date']
```

```python
# To prevent biased ratings,Drop games that has rating less than 200 user rating count AND 
# (Release Date to Update date) less than 6 months
low_rating_count = (df3[df3['User Rating Count'] < 200].index) & (df3[df3['Update Gap'] < datetime.timedelta(days=175)].index)

df3.drop(low_rating_count, inplace=True)
```

```python
# Create another for further modifications
df4 = df3.copy()
```

```python
df4['Genres'] = df4['Genres'].str.replace(',', '').str.replace('Games', '').str.replace('Entertainment', '').str.replace('Strategy', '')
df4['Genres'] = df4['Genres'].str.split(' ').map(lambda x: ' '.join(sorted(x)))
df4['Genres']=df4['Genres'].str.strip()
```

```python
Non_Main_Genre=df4[~df4.Genres.str.contains('Puzzle') & \
                           ~df4.Genres.str.contains('Action') & \
                           ~df4.Genres.str.contains('Family')& \
                           ~df4.Genres.str.contains('Education')& \
                           ~df4.Genres.str.contains('Family')& \
                           ~df4.Genres.str.contains('Adventure')&\
                           ~df4.Genres.str.contains('Board')&\
                           ~df4.Genres.str.contains('Role')].index
df4.drop(Non_Main_Genre , inplace=True)
df4.loc[df4['Genres'].str.contains('Puzzle'),'Genres'] = 'Puzzle'
df4.loc[df4['Genres'].str.contains('Board'),'Genres'] = 'Puzzle'
df4.loc[df4['Genres'].str.contains('Action'),'Genres'] = 'Action'
df4.loc[df4['Genres'].str.contains('Adventure'),'Genres'] = 'Adventure'
df4.loc[df4['Genres'].str.contains('Role'),'Genres'] = 'Adventure'
df4.loc[df4['Genres'].str.contains('Family'),'Genres'] = 'Family'
df4.loc[df4['Genres'].str.contains('Education'),'Genres'] = 'Family'
```

```python
print(df4.info())
df4.sample(5)
```

## Create a CSV file of the cleaned data

```python
df4.to_csv('games.csv')
```

```python
df5=df4.copy()
import statistics
df5['In-app Purchases'] = df5['In-app Purchases'].str.replace(',', '') 
df5['In-app Purchases'] = df5['In-app Purchases'].fillna(0)

for i in (list(df5.index.values)):
    if df5['In-app Purchases'][i] ==0:
        df5['In-app Purchases'][i]=0.01
    elif df5['In-app Purchases'][i] !=0:
        A=str.split(df5['In-app Purchases'][i])
        A=[float(i) for i in A]
        df5['In-app Purchases'][i]=statistics.mean(A)
vis1= sns.stripplot(y=df5['In-app Purchases'] , x=df5.Price, data=df5, hue=df5.Genres, dodge=True, size=3)

vis1.set(xlabel='Price in $',ylabel='Average in App Purchase in $')
vis1.set(xlim=(-0.7, 12))
plt.ioff()
```

# Conclusion

1. For simpler games, focus on size between 100MB to 150MB of contents
2. For more complex games, target a content of 500MB and above
3. Games above 1GB have good tendancy to land above 3 score and above, provided sufficient resource available to the developer
4. Puzzle games are less saturated above 400MB to achieve score 4.0 and above
5. Action & Adventure Genre games are less saturated above 600MB to achieve score 4.0 and above
6. Family/Educational games has less competition overall
7. Almost 90% of the developers focuses on games below 1 Dollar and places In-App Purchase as their strategy for income
8. On average, the trend shows that the average game size increase of 30MB per/year but small size games are still relatable in the store.
9. Puzzle and Family games requires less game updates on average.
10. Paid Games tend to have better reviews espeicially those above 9 Dollar scores 4.0 User Rating
11. Consumers enjoy more matured content i.e games targetted for 17+ age and above
